```python
# 구구단 출력
num = ""
while num != 1 :
    num = int(input("구구단 몇 단을 출력할까요? (종료 1) : "))
    if num == 1 :
        print("종료합니다.")
    else :
        for i in range(1,10) :
            print(str(num)+' * '+str(i)+' = '+str(num*i))
    
```

```python
# 숫자 맞추기
answer = int(input("정답 입력하기 : "))
num = 0
while num != answer :
    num = int(input("예상 숫자를 입력하세요 : "))
    if num > answer :
        print("DOWN")
    elif num < answer :
        print("UP")
    else :
        print("정답!")
```

```python
s = [80, 85, 95, 99, 60, 55]
n = 0
print("한국사 시험 결과 : ")
for i in s :
    n = n + 1
    if i >= 70 :
        print(n, "번 학생은 1급입니다.")
    elif i >= 60 :
        print(n, "번 학생은 2급입니다.")
    else :
        print(n,"번 학생은 불합격입니다.")
```

```python
a = int(input("홀수는 1, 짝수는 2를 입력해주세요 : "))
for i in range(a, 101, 2) :
    print(i)
```

```python
num1 = int(input("마지막 숫자는 : "))
num2 = int(input("뛰어넘을 숫자는 : "))

for i in range(num1 + 1) :
    if i == num2 :
        continue
    print(i)
```

```python
print("구구단 만들기~~")
for i in range(2, 10) :
    print("\n")
    print(i, "단 !")
    for j in range(1, 10) :
        print(i, "*", j, "=", i*j)
```
