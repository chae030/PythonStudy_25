# list

```python
a = ['a', 'b', 'c', 'd', 'e']
del a[0:2]
print(a)

a.append('f')
print(a)

a.insert(2, 'g')
print(a)

a.extend(['a', 'b'])
print(a)

print(a.count('c'))
print(a.index('a'))

a.remove('e')
print(a)
```

```python
a = [3, 1, 4, 5, 2]
print(a)

a.sort()
print(a)

a.reverse()
print(a)
```

```python
st = ["이황", "이이", "원효"]
print(f"현재 학생은 {st}")
a = input("전학 온 학생은 누구입니까? : ")
st.append(a)
print(st)
print("번호를 정렬합니다.")
st.sort()
b = 0
for i in st :
    b = b + 1
    print(b, i)
```

```python
# 튜플 패킹, 언패킹

pack = 1, 2, 3, 4, 5, 4, 3, 2, 1
a, b, *c = pack

print(f"a는 {a} b는 {b} c는 {c}")

a, b = b, a
print(f"a는 {a} b는 {b}")
```

# set
mutable
순서 없음

```python
a = [1, 3, 2, 2, 5, 4, 7, 6]
print(set(a))
```

```python
a = set(range(1, 6))
b = set(range(3, 8))
c = set(range(1, 3))

print(a|b, a.union(b)) # 합집합
print(a&b, a.intersection(b)) # 교집합
print(a-b, a.difference(b)) # 차집합
print(a^b, a.symmetric_difference(b)) # 대칭차집합
print(c <= a, c.issubset(a)) # 부분집합 검사하기
print(c >= a, c.issuperset(a)) # 상위집합 검사하기
print(c.isdisjoint(b)) # 서로소 검사하기
```

```python
a = set(range(1, 6))
print(a)

a.pop()
print(a)
a.pop()
print(a)

a = [1, 2, 3, 4, 5]
a.pop()
print(a)
```

```python
a = set(range(1, 6))
print(a.pop())
print(a)

# pop 은 임의의 요소를 반환
```

```python
a = set(range(1, 6))
a.discard(3)
print(a)
a.add(6)
print(a)
a.update({7, 8})
print(a)
print(max(a), min(a))
a.clear()
print(a)
```

# dictionary
key 변경 불가 value 변경 가능

```python
a = {1 : 'apple', 2 : 'banana', 'three' : 3333, 'four' : '4444'}
print(a[2])
print(a['three'])

a['Alexis'] = 2
print(a)

print(2 in a)
print(3 in a)
print('four' in a)

print(a.pop("three"))

print(a.keys())
print(a.items())
print(a.values())
```
