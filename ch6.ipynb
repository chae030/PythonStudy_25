```python
# practice 1 인공지능 기초 (해당 키워드에 대한 답을 저장해두고 출력해보기)

while True :
    order = input("명령 : ")
    if "천재" in order :
        print("고맙핑~~ 너도천재얌")
    elif "바보" in order :
        print("너가 더 바보임")
    elif "에잇" in order : 
        print("에잇! 그만하자.")
    elif "종료" in order :
        print("종료합니다.")
        break
```

```python
while True :
    a = input("비밀번호를 입력하세요 : ")
    if len(a) < 8 :
        print("8자리 이상을 입력하세요.")
    elif a.isalpha() == True or a.isdigit() == True :
        print("영문자와 숫자의 조합으로 입력하세요.")
    else :
        print("비밀번호는 %s 입니다." %a)
        break
```
